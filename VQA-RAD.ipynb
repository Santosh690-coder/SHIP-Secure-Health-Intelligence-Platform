```python
from datasets import load_dataset
import os

dataset = load_dataset("flaviagiammarino/vqa-rad")

os.makedirs("unique_vqa_images", exist_ok=True)

saved = set()
count = 0

for row in dataset["train"]:
    img = row["image"]
    img_bytes = img.tobytes()

    if img_bytes not in saved:
        saved.add(img_bytes)
        img.save(f"unique_vqa_images/image_{count}.jpg", "JPEG")
        count += 1

print("Total unique JPG images:", count)
```
