```python
%pip install -q kagglehub

import os
import shutil
import kagglehub
from pathlib import Path

BASE_DIR = "AI_Doctor_Datasets"
os.makedirs(BASE_DIR, exist_ok=True)

print(f"Main Folder Created: {BASE_DIR}")

def collect_images(dataset_path, output_dir):
    
    # Supported image extensions
    image_extensions = (
        ".png", ".jpg", ".jpeg", ".bmp", ".tiff", ".webp"
    )

    os.makedirs(output_dir, exist_ok=True)

    image_count = 0

    # Walk through all folders
    for root, dirs, files in os.walk(dataset_path):
        for file in files:
            if file.lower().endswith(image_extensions):
                source_path = os.path.join(root, file)

                # Create unique filename
                new_filename = f"img_{image_count}{Path(file).suffix}"

                destination_path = os.path.join(
                    output_dir,
                    new_filename
                )

                shutil.copy2(source_path, destination_path)
                image_count += 1

    print(f"\nCollected {image_count} images")
    print(f"Saved to: {output_dir}")

print("\n======================================")
print("DOWNLOADING DATASET 1")
print("Synthetic Medical Prescription OCR")
print("======================================")

dataset1_path = kagglehub.dataset_download(
    "priyanshuyav13/synthetic-medical-prescription-ocr-dataset"
)

print("Downloaded Dataset 1 Path:")
print(dataset1_path)

dataset1_output = os.path.join(
    BASE_DIR,
    "Medical_Prescription_Images"
)

# Corrected function call
collect_images(dataset1_path, dataset1_output)

print("\n======================================")
print("DOWNLOADING DATASET 2")
print("Illegible Medical Prescription Images")
print("======================================")

dataset2_path = kagglehub.dataset_download(
    "mehaksingal/illegible-medical-prescription-images-dataset"
)

print("Downloaded Dataset 2 Path:")
print(dataset2_path)

dataset2_output = os.path.join(
    BASE_DIR,
    "Illegible_Prescription_Images"
)

# Corrected function call
collect_images(dataset2_path, dataset2_output)

print("\n======================================")
print("DOWNLOAD COMPLETE")
print("======================================")

print(f"""
Datasets Ready:

1. Medical Prescription Images
   Folder:
   {dataset1_output}

2. Illegible Prescription Images
   Folder:
   {dataset2_output}
""")
```

```python
import hashlib

def remove_duplicate_images(folder_path):

    hashes = {}
    duplicate_count = 0

    for root, dirs, files in os.walk(folder_path):

        for file in files:

            file_path = os.path.join(root, file)

            try:
                with open(file_path, "rb") as f:
                    file_hash = hashlib.md5(f.read()).hexdigest()

                # Duplicate found
                if file_hash in hashes:
                    os.remove(file_path)
                    duplicate_count += 1

                else:
                    hashes[file_hash] = file_path

            except Exception as e:
                print(f"Error processing {file}: {e}")

    print(f"\nRemoved {duplicate_count} duplicate images")
```
