CSV format with ~16.4k rows.

```python
from datasets import load_dataset
import pandas as pd

# Fast load from Hugging Face
dataset = load_dataset("keivalya/MedQuad-MedicalQnADataset")

# Convert directly to pandas DataFrame (fast)
df = dataset["train"].to_pandas()

# Save CSV
df.to_csv("Cleaned_medquad_dataset.csv", index=False)

print("CSV saved successfully!")
print("Shape:", df.shape)
print(df.head())
```

DataFrame into JSON

```python
from datasets import load_dataset

# Load dataset
dataset = load_dataset("keivalya/MedQuad-MedicalQnADataset")

# Convert to pandas
df = dataset["train"].to_pandas()

# Keep required columns
df = df[["Question", "Answer"]]

# Convert directly to JSON file (fast)
df.to_json("Cleaned_medquad.json", orient="records", indent=4)

print("JSON file saved successfully!")
print(df.head())
```
