```python
from datasets import load_dataset
import os

dataset = load_dataset(
    "eltorio/ROCOv2-radiology",
    split="train",
    streaming=True
)

save_folder = "roco_train_images"
os.makedirs(save_folder, exist_ok=True)

for i, row in enumerate(dataset):
    if i == 1000:
        break

    row["image"].save(f"{save_folder}/image_{i}.jpg")

    if i % 100 == 0:
        print(f"Saved {i} images...")

print("Done")
```

```python
# Separate folder for 1000 JPG images
save_folder = "roco_1000_jpg_images"
os.makedirs(save_folder, exist_ok=True)

# Save exactly 1000 images
for i, row in enumerate(dataset):
    if i >= 1000:
        break

    img = row["image"]
    file_path = os.path.join(save_folder, f"image_{i}.jpg")
    img.save(file_path, format="JPEG")

    if (i + 1) % 100 == 0:
        print(f"{i + 1} images saved")

print("Done!")
print("Folder name:", save_folder)
print("Total images:", len(os.listdir(save_folder)))
```
