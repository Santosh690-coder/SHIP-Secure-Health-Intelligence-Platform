```python
import json
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings

print("1. Loading Verified Medical Facts for RAG...")
# We will use MedQuad since it's pure, high-quality medical factual Q&A
with open("Cleaned_medquad.json", "r", encoding="utf-8") as f:
    medquad_data = json.load(f)

# Combine the questions and answers into searchable documents
# (We are taking the first 10,000 to keep it fast and fit in your 16GB RAM)
print("2. Formatting text documents...")
texts = []
for item in medquad_data[:10000]:
    texts.append(f"Medical Concept: {item['Question']}\nFactual Answer: {item['Answer']}")

print("3. Downloading Embedding Model (This converts text to searchable math)...")
# This is a very tiny, fast model perfect for your system
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

print("4. Building the FAISS Vector Database (This may take a few minutes)...")
vector_db = FAISS.from_texts(texts, embeddings)

# Save the database to your hard drive so we never have to build it again!
vector_db.save_local("medical_vector_db")

print("\n✅ RAG Database Built and Saved to your PC!")
```

```python
from datasets import load_dataset
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings

print("1. Downloading PubMed QA (Verified Medical Research)...")
# PubMed QA contains verified medical research contexts and expert conclusions
pubmed_dataset = load_dataset("pubmed_qa", "pqa_labeled", split="train")

print("2. Formatting text documents for the AI to read...")
texts = []
for item in pubmed_dataset:
    # We combine the background medical context with the final verified conclusion
    context = " ".join(item['context']['contexts'])
    conclusion = item['long_answer']
    texts.append(f"Medical Context: {context}\nConclusion: {conclusion}")

print(f"Prepared {len(texts)} highly verified medical documents.")

print("3. Loading Embedding Model (Converting text to searchable math)...")
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

print("4. Building the FAISS Vector Database (This may take a few minutes)...")
# We will use the first 1000 items to keep it light on your 16GB RAM
vector_db = FAISS.from_texts(texts[:1000], embeddings)

# Save the new database!
vector_db.save_local("pubmed_medical_vector_db")

print("\n✅ High-Quality PubMed RAG Database Built and Saved!")
```
