```python
import json
import os 
folder = "symptoms_diagnosis_dataset"
# Load all datasets
with open("Cleaned_medquad.json", "r", encoding="utf-8") as f:
    medquad = json.load(f)

with open("chatdocter_Cleaned_data.json", "r", encoding="utf-8") as f:
    chatdoctor = json.load(f)

with open("medical_meadow_dataset.json", "r", encoding="utf-8") as f:
    medical_meadow = json.load(f)

with open("Cleaned_Symptoms.json", "r", encoding="utf-8") as f:
    symptoms = json.load(f)

merged_data = []

# 1. MedQuad
for row in medquad:
    merged_data.append({
        "source": "MedQuad",
        "input": row.get("Question", ""),
        "output": row.get("Answer", "")
    })

# 2. Chat Doctor
for row in chatdoctor:
    merged_data.append({
        "source": "ChatDoctor",
        "input": row.get("input", ""),
        "output": row.get("output", "")
    })

# 3. Medical Meadow
for row in medical_meadow:
    text_input = row.get("instruction", "") + " " + row.get("input", "")
    
    merged_data.append({
        "source": "MedicalMeadow",
        "input": text_input.strip(),
        "output": row.get("output", "")
    })

# 4. Symptoms to Diagnosis
for row in symptoms:
    merged_data.append({
        "source": "SymptomsDiagnosis",
        "input": row.get("diagnosis", ""),
        "output": row.get("text", "")
    })

# Save merged dataset
with open("merged_medical_dataset.json", "w", encoding="utf-8") as f:
    json.dump(merged_data, f, indent=4, ensure_ascii=False)

print("Merged successfully")
print("Total rows:", len(merged_data))
```

```python
unique_data = []
seen = set()

for row in merged_data:
    pair = (row["input"], row["output"])
    if pair not in seen:
        seen.add(pair)
        unique_data.append(row)

print("After duplicate removal:", len(unique_data))
```

```python
with open("clean_merged_dataset.json", "w", encoding="utf-8") as f:
    json.dump(unique_data, f, indent=4)
```
