KEEP duplicate columns (advanced)

```python
def make_unique_columns(df):
    cols = pd.Series(df.columns)
    for dup in cols[cols.duplicated()].unique():
        cols[cols[cols == dup].index] = [
            f"{dup}_{i}" if i != 0 else dup
            for i in range(sum(cols == dup))
        ]
    df.columns = cols
    return df

df1 = make_unique_columns(df1)
df2 = make_unique_columns(df2)
df3 = make_unique_columns(df3)
df4 = make_unique_columns(df4)
```

```python
import pandas as pd
import json
import uuid

print("Starting Instruction Formatting...")

final_training_data = []

# ==========================================
# 1. Format the Image Dataset
# ==========================================
print("Processing Image Data...")
image_df = pd.read_csv("final_master_qa_dataset.csv")

for index, row in image_df.iterrows():
    # Make sure we don't have empty rows
    if pd.isna(row['question']) or pd.isna(row['answer']):
        continue
        
    final_training_data.append({
        "id": f"img_{uuid.uuid4().hex[:8]}", # Generate a unique ID
        "image": str(row['image_identifier']),
        "conversations": [
            # Multimodal models usually need an <image> token in the prompt
            {"from": "human", "value": f"<image>\n{row['question']}"}, 
            {"from": "gpt", "value": str(row['answer'])}
        ]
    })

# ==========================================
# 2. Format the Text/Chat Dataset
# ==========================================
print("Processing Text Data...")
# Adjust the path if yours is saved differently!
text_df = pd.read_json(r"D:\Santosh D\Study\College\Projects\SHIP\together_final_merged.json")

for index, row in text_df.iterrows():
    # Get input/output (you named these columns in Untitled12.ipynb)
    user_input = row.get('input', '')
    ai_output = row.get('output', '')
    
    if pd.isna(user_input) or pd.isna(ai_output) or user_input == '' or ai_output == '':
        continue
        
    final_training_data.append({
        "id": f"txt_{uuid.uuid4().hex[:8]}",
        # Notice: No "image" key here!
        "conversations": [
            {"from": "human", "value": str(user_input)},
            {"from": "gpt", "value": str(ai_output)}
        ]
    })

# ==========================================
# 3. Save the Final AI Training File
# ==========================================
print("Saving Final Dataset...")
output_file = "AI_Training_Dataset.json"
with open(output_file, "w", encoding="utf-8") as f:
    json.dump(final_training_data, f, indent=4)

print(f"🎉 Success! Combined {len(final_training_data)} total training examples.")
print(f"File saved as: {output_file}")
```
