```python
%pip install -q datasets opencv-python-headless pandas pillow

import numpy as np
import os
import cv2
import pandas as pd

from datasets import load_dataset

output_folder = "Handwritten_Medical_Records"

os.makedirs(output_folder, exist_ok=True)

dataset = load_dataset(
    "chaithanyakota/100-handwritten-medical-records"
)

image_paths = []

for i, item in enumerate(dataset["train"]):

    image = item["image"]

    # Convert PIL image to numpy
    img = np.array(image)

    # If RGB image -> convert to grayscale
    if len(img.shape) == 3:
        img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

    # Resize
    img = cv2.resize(img, (512, 512))

    # Denoise
    img = cv2.fastNlMeansDenoising(img)

    # OCR preprocessing
    processed = cv2.adaptiveThreshold(
        img,
        255,
        cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
        cv2.THRESH_BINARY,
        11,
        2
    )

    # Save image
    image_path = os.path.join(
        output_folder,
        f"record_{i}.png"
    )

    cv2.imwrite(image_path, processed)

    image_paths.append(image_path)

print("Total Images Saved:", len(image_paths))

df = pd.DataFrame({
    "image_path": image_paths,
    "text": ["" for _ in image_paths]
})

df.to_csv(
    "medical_records_labels.csv",
    index=False
)

print("CSV Created")
```
